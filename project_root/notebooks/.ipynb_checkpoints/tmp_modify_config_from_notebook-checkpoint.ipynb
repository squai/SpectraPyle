```python
from copy import deepcopy
from spectraPyle.stacking.stacking import main, load_config_file

base_cfg = load_config_file("../configs/YAML/default_xavi.yaml")

print (base_cfg)
```

```python
main(config=base_cfg)
```

```python
file_names = ["file_1", "file_2"]
spectra_datafile = ["spectra_file_1", "spectra_file_2"]   
                  
for name, spec in zip(file_names, spectra_datafile):
    cfg = deepcopy(base_cfg) 

    cfg["io"]["filename_in"] = name
    cfg["io"]["spectra_datafile"] = spec

    print ('-------')
    print (cfg)
    main(config=cfg)
    
    
```

```python
file_names = ["best_inspected"]
file_extention = ["csv"]
spectra_datafile = ["SPECTRA_COMBINED_galaxy"]   
                  
for name, ext, spec in zip(file_names, file_extention, spectra_datafile):
    cfg = deepcopy(base_cfg) 

    cfg["io"]["filename_in"] = name
    cfg["io"]["filename_in_extention"] = ext
    cfg["io"]["spectra_datafile"] = spec
    
    cfg["catalog_columns"]["ID_column_name"] = "object_id"
    cfg["catalog_columns"]["redshift_column_name"] = "spe_z"
    
    
    print ('-------')
    print (name, ext, spec)
    print (cfg)
    main(config=cfg)
    
    
```
