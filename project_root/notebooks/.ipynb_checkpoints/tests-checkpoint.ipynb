```python
import numpy as np
from scipy.stats import binned_statistic
```

```python

def resamplingSpecFluxCons(lbd, flux, lambdaInterp):
    """
    Resamples a step-function spectrum (lbd, flux) onto a new grid lambdaInterp while conserving flux.
    
    Parameters:
    - lbd: Original wavelength array (pixel centers)
    - flux: Original flux array
    - lambdaInterp: New wavelength grid (bin edges, length N+1)

    Returns:
    - flux_resampled: The resampled flux array (length N)
    """
    
    # Step 1: Compute pixel edges from pixel centers
    lbd_edges = np.zeros(len(lbd) + 1)
    lbd_edges[1:-1] = 0.5 * (lbd[1:] + lbd[:-1])
    lbd_edges[0] = lbd[0] - 0.5 * (lbd[1] - lbd[0])
    lbd_edges[-1] = lbd[-1] + 0.5 * (lbd[-1] - lbd[-2])
    
    # Step 2: Construct fine sampling within each original pixel
    fine_dx = 0.01  # 0.01 Angstrom resolution

    x_fine = []
    y_fine = []

    for i in range(len(flux)):
        start, end = lbd_edges[i], lbd_edges[i+1]
        x_seg = np.arange(start, end, fine_dx)
        if len(x_seg) == 0:
            x_seg = np.array([(start + end) / 2])  # fallback
        x_fine.append(x_seg)
        y_fine.append(np.full_like(x_seg, flux[i]))

    x_fine = np.concatenate(x_fine)
    y_fine = np.concatenate(y_fine)
    
    # Step 3: Integrate within new bins
    flux_integrated, _, _ = binned_statistic(x_fine, y_fine, statistic="sum", bins=lambdaInterp)
    dx_fine = np.mean(np.diff(x_fine))
    
    # Step 4: Normalize to step function
    bin_widths = np.diff(lambdaInterp)
    flux_resampled = flux_integrated * dx_fine / bin_widths

    # Optional: Mask out bins outside original range
    min_lbd = lbd_edges[0]
    max_lbd = lbd_edges[-1]
    flux_resampled[(lambdaInterp[:-1] < min_lbd) | (lambdaInterp[:-1] > max_lbd)] = np.inf
    
    return flux_resampled

import numpy as np

def resamplingSpecFluxCons_drizzle_1d(lbd, flux, var, lambdaInterp):
    """
    Fully vectorized 1D drizzle with variance propagation. Adapted from Fruchter & Hook 2002
    Produces flux density per unit wavelength, consistent with the non-vectorized version.

    Parameters
    ----------
    lbd : (N_in,) array
        Input wavelength pixel centers
    flux : (N_in,) array
        Flux density per input pixel
    var : (N_in,) array
        Variance per input pixel
    lambdaInterp : (N_out+1,) array
        Output wavelength bin edges

    Returns
    -------
    flux_out : (N_out,) array
        Drizzled flux density (per unit wavelength)
    err_out : (N_out,) array
        Drizzled error (per unit wavelength)
    """

    # --- Input pixel edges ---
    lbd_edges = np.zeros(len(lbd) + 1)
    lbd_edges[1:-1] = 0.5 * (lbd[1:] + lbd[:-1])
    lbd_edges[0] = lbd[0] - 0.5 * (lbd[1] - lbd[0])
    lbd_edges[-1] = lbd[-1] + 0.5 * (lbd[-1] - lbd[-2])

    delta_p = np.diff(lbd_edges)                      # Width of input pixels
    p_lo = lbd_edges[:-1][None, :]                    # Shape (1, N_in)
    p_hi = lbd_edges[1:][None, :]                     # Shape (1, N_in)

    # --- Output bins ---
    j_lo = lambdaInterp[:-1][:, None]                 # Shape (N_out, 1)
    j_hi = lambdaInterp[1:][:, None]                  # Shape (N_out, 1)
    delta_j = (j_hi - j_lo).flatten()                # Output bin widths, shape (N_out,)

    # --- Fractional overlaps ---
    overlap = np.maximum(0.0, np.minimum(p_hi, j_hi) - np.maximum(p_lo, j_lo))  # (N_out, N_in)

    # --- Weighted sums ---
    flux_sum = overlap @ flux                          # (N_out,)
    var_sum  = (overlap**2) @ var                     # (N_out,)

    # --- Divide by output bin width to get flux density per unit wavelength ---
    flux_out = flux_sum / delta_j
    err_out  = np.sqrt(var_sum) / delta_j

    return flux_out, err_out


def drizzle_1d_flux_var(lbd, flux, var, lambdaInterp):
    """
    1D flux-conserving drizzle with variance propagation (corrected).

    Parameters
    ----------
    lbd : array
        Input wavelength pixel centers
    flux : array
        Flux density per input pixel
    var : array
        Variance per input pixel
    lambdaInterp : array
        Output wavelength bin edges (length N+1)

    Returns
    -------
    flux_out : array
        Drizzled flux density per output bin (length N)
    var_out : array
        Drizzled variance per output bin (length N)
    """

    # Input pixel edges
    lbd_edges = np.zeros(len(lbd) + 1)
    lbd_edges[1:-1] = 0.5 * (lbd[1:] + lbd[:-1])
    lbd_edges[0] = lbd[0] - 0.5 * (lbd[1] - lbd[0])
    lbd_edges[-1] = lbd[-1] + 0.5 * (lbd[-1] - lbd[-2])

    n_out = len(lambdaInterp) - 1
    flux_out = np.zeros(n_out)
    var_out = np.zeros(n_out)

    for j in range(n_out):
        lo, hi = lambdaInterp[j], lambdaInterp[j+1]

        # Compute overlap of each input pixel with output bin
        overlap_lo = np.maximum(lbd_edges[:-1], lo)
        overlap_hi = np.minimum(lbd_edges[1:], hi)
        overlap = np.maximum(0.0, overlap_hi - overlap_lo)  # absolute overlap in wavelength units

        # Drizzled flux density (flux per unit wavelength)
        flux_out[j] = np.sum(flux * overlap) / (hi - lo)
        var_out[j]  = np.sum(var * overlap**2) / (hi - lo)**2

    return flux_out, var_out
```

```python
def compare_resampling_with_variance(lbd, flux, var, lambda_edges_out):
    """
    Compare fine-grid resampling and drizzle including variance.
    """

    flux_fine = resamplingSpecFluxCons(lbd, flux, lambda_edges_out)

    # Approximate variance for fine-grid method
    flux_fine_err = resamplingSpecFluxCons(lbd, np.sqrt(var), lambda_edges_out)

    flux_drizzle, err_drizzle = resamplingSpecFluxCons_drizzle_1d(
        lbd, flux, var, lambda_edges_out
    )

    return {
        "flux_fine": flux_fine,
        "err_fine": flux_fine_err,
        "flux_drizzle": flux_drizzle,
        "err_drizzle": err_drizzle
    }
```

```python
def compare_resampling_with_variance_v2(lbd, flux, var, lambda_edges_out):
    """
    Compare fine-grid resampling and drizzle including variance.
    """

    flux_fine = resamplingSpecFluxCons(lbd, flux, lambda_edges_out)

    # Approximate variance for fine-grid method
    flux_fine_err = resamplingSpecFluxCons(lbd, np.sqrt(var), lambda_edges_out)

    flux_drizzle, err_drizzle = drizzle_1d_flux_var(
        lbd, flux, var, lambda_edges_out
    )

    return {
        "flux_fine": flux_fine,
        "err_fine": flux_fine_err,
        "flux_drizzle": flux_drizzle,
        "err_drizzle": err_drizzle
    }
```

```python
def make_test_spectrum():
    lbd = np.linspace(4000, 5000, 800)

    continuum = (lbd / 4500)**-1.5
    line1 = np.exp(-(lbd - 4300)**2 / (2*2.0**2))
    line2 = 0.7 * np.exp(-(lbd - 4700)**2 / (2*4.0**2))

    flux = continuum + line1 + line2
    return lbd, flux
```

```python
lbd, flux = make_test_spectrum()

# Toy variance: S/N ~ 20
var = (flux / 20.0)**2

lambda_edges_out = np.arange(4020, 4980, 0.75)

result = compare_resampling_with_variance(
    lbd, flux, var, lambda_edges_out
)

result_v2 = compare_resampling_with_variance_v2(
    lbd, flux, var, lambda_edges_out
)
```

```python
# --- logarithmic sampling ---
pixel_resampling = 1e-3
wavelMin = 9000
wavelMax = 18000

dloglam = pixel_resampling   # interpreted as Δlog10(λ)

loglam_min = np.log10(wavelMin)
loglam_max = np.log10(wavelMax)

npix = int(np.round((loglam_max - loglam_min) / dloglam))
loglam_max = loglam_min + npix * dloglam

loglam_bins = loglam_min + np.arange(npix + 1) * dloglam
loglam_centers = loglam_bins[:-1] + 0.5 * dloglam

wavelength_bins = 10.0 ** loglam_bins
wavelength_centers = 10.0 ** loglam_centers
```

```python
def dloglam_from_R(R, nsamp=5):
    """
    Compute logarithmic wavelength step Δlog10(λ) from resolving power.

    Parameters
    ----------
    R : float
        Spectral resolving power (λ / Δλ_FWHM)
    nsamp : float, optional
        Number of pixels per resolution element (default: 5)

    Returns
    -------
    dloglam : float
        Logarithmic wavelength step Δlog10(λ)
    """
    if R <= 0:
        raise ValueError("R must be positive")
    if nsamp <= 0:
        raise ValueError("nsamp must be positive")

    return 1.0 / (R * nsamp * np.log(10.0))


def dlam_from_R(R, lambda_ref, nsamp=5):
    """
    Δλ (Å) from resolving power at a reference wavelength.

    Parameters
    ----------
    R : float
        Spectral resolving power (λ / Δλ_FWHM)
    lambda_ref : float
        Reference wavelength (Å)
    nsamp : float, optional
        Number of pixels per resolution element (default: 5)

    Returns
    -------
    dlam : float
        Linear wavelength step (Å)
    """
    if R <= 0:
        raise ValueError("R must be positive")
    if nsamp <= 0:
        raise ValueError("nsamp must be positive")
    if lambda_ref <= 0:
        raise ValueError("lambda_ref must be positive")

    return lambda_ref / (R * nsamp)
```

```python
import h5py

def print_hdf5_structure(name, obj, level=0):
    """Pretty-print HDF5 structure recursively."""
    prefix = '  ' * level
    if isinstance(obj, h5py.File) or isinstance(obj, h5py.Group):
        print(f"{prefix}{name}/")
        for key in obj.keys():
            print_hdf5_structure(key, obj[key], level+1)
    else:
        shape = obj.shape if hasattr(obj, 'shape') else 'scalar'
        dtype = obj.dtype
        print(f"{prefix}{name} (shape={shape}, dtype={dtype})")

# Usage
with h5py.File(f"{namein}.h5", 'r') as f:
    print_hdf5_structure('file', f)
```
